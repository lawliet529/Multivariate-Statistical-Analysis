# Tuần 3: Ellipsoid

## 1. Sinh ngẫu nhiên 1200 giá trị của biến X biết X ~ N(165, 21.2) và 1200 giá trị của biến Y biết Y ~ N(175, 25.5). Vẽ mặt mức tương ứng với vector ngẫu nhiên 2-chiều

```r
library(ellipse)
library(MASS)
library(mnormt)
```

```r
x <- rnorm(1200, 165, 21.2)
y <- rnorm(1200, 175, 25.2)
data <- data.frame(x, y)
S <- cov(data)
cm <- colMeans(data)
plot(ellipse(S, centre = cm))
```

## 2. Cho vector ngẫu nhiên 2-chiều. Xét một mẫu có kỳ vọng mẫu; var(X) = 21.2; var(Y) = 25.5; cov(X,Y) = 0. Dùng hàm mvrnorm sinh ngẫu nhiên 1000 giá trị của vecto 2-chiều đó. Kiểm định xem từng biến ngẫu nhiên trong đó có phân phối chuẩn không?

```r
S <- matrix(c(1, 0, 0, 1), nrow = 2)
cm <- c(21.2, 25.5)
X <- mvrnorm(1000, cm, S)
```

```r
shapiro.test(X[, 1])
shapiro.test(X[, 2])
```

```r
# p-value > 0.05
# => X có phân bố chuẩn
```

## 3. Cho vector ngẫu nhiên 2-chiều. Xét một mẫu của nó, trong đó X gồm các giá trị từ -10 đến 10, các giá trị liên tiếp cách nhau 0.1 và Y gồm các giá trị từ -10 đến 30, các giá trị liên tiếp cách nhau 0.2. Vẽ đồ thị các đường đồng mức tương ứng.

```r
x <- seq(-10, 10, 0.1)
y <- seq(-10, 30, 0.2)
mu <- c(0, 0)
sigma <- matrix(c(1, 0, 0, 1), nrow = 2)
```

```r
f <- function(x, y) dmnorm(cbind(x, y), mu, sigma)
z <- outer(x, y, f)
```

```r
contour(x, y, z)
```
